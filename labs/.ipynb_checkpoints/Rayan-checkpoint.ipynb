```python
def neighboring_pixels(img, pixel, treshold):
    """"
    This method get the intensity of the neighboring pixel around a seed. It is able to process any pixel of 
    an image included the frame. 
    -----------------------------------------------------------------------------------------------------------
    Inputs: 
    img: an array containing an image
    pixel: the coordinates of a specific pixel
    treshold: value under which the pixel doesn't belong to the generated region
    ----------------------------------------------------------------------------------------------------------- 
    Return:
    - a list of the remaining coordinate 
    """
    
    #x=pixel[0]
    #y=pixel[1]
    (x,y) = pixel
    neighbors = []
    
    # computing for the inner img
    if x>0 or x<im_w:
        pixel_ml=img[x-1][y]    # middle left
        pixel_mr=img[x+1][y]
        pixel_tl=img[x-1][y-1]
        pixel_tr=img[x+1][y-1]  # top right
        pixel_bl=img[x-1][y+1]
        pixel_br=img[x+1][y+1]
        pixel_tm=img[x][y-1]
        pixel_bm=img[x][y+1]    # bottom middle
        
    # computing for the frame
    if x==0:
        pixel_ml=0
        pixel_tl=pixel_bl=pixel_ml

        pixel_mr=img[x+1][y]
        pixel_tr=img[x+1][y-1]
        pixel_br=img[x+1][y+1]
        pixel_tm=img[x][y-1]
        pixel_bm=img[x][y+1]
        
        if y==0:
            pixel_tm=0
            pixel_tr=pixel_tm
                
        if y==im_h:
            pixel_bm=pixel_br=0
    if x==im_w:
        pixel_mr=0
        pixel_tr=pixel_br=pixel_mr
        
        pixel_ml=img[x-1][y]
        pixel_tl=img[x-1][y-1]
        pixel_bl=img[x-1][y+1]
        pixel_tm=img[x][y-1]
        pixel_bm=img[x][y+1]
        
        if y==0:
            pixel_tm=pixel_tl=0
            
        if y==im_h:
            pixel_bm=pixel_bl=0
    
    if (x>0 or x<im_w) and y==0:
        pixel_tl=pixel_tm=pixel_tr=0
        
        pixel_ml=img[x-1][y]
        pixel_mr=img[x+1][y]
        pixel_bl=img[x-1][y+1]
        pixel_br=img[x+1][y+1]
        pixel_bm=img[x][y+1]
        
    if (x>0 or x<im_w) and y==im_h:
        pixel_bl=pixel_bm=pixel_br=0
        
        pixel_ml=img[x-1][y]
        pixel_mr=img[x+1][y]
        pixel_tl=img[x-1][y-1]
        pixel_tr=img[x+1][y-1]
        pixel_tm=img[x][y-1]
        
    # Every element is added according to a specific order: top to bottom and left to right
    
    if pixel_tl > treshold:
        neighbors.append((x-1,y-1))
    if pixel_tm > treshold:
        neighbors.append((x,y-1))
    if pixel_tr > treshold:
        neighbors.append((x+1,y-1))
    if pixel_ml > treshold:
        neighbors.append((x-1,y))
    if pixel_mr > treshold:
        neighbors.append((x+1,y))
    if pixel_bl > treshold:
        neighbors.append((x-1,y+1))
    if pixel_bm > treshold:
        neighbors.append((x,y+1))
    if pixel_br > treshold:
        neighbors.append((x+1,y+1))
       
    return neighbors
```

```python
def region_growing(img, seed, threshold):
    """"
    This method grows a region according to a seed selected in the region of interest and a threshold value. 
    -----------------------------------------------------------------------------------------------------------
    Inputs: 
    img: an array containing an image.
    seed: the coordinate of the pixel belonging to the region of interest.
    ----------------------------------------------------------------------------------------------------------- 
    Return:
    - an image of the desired region
    """
    import numpy as np
    
    # definition of the returned image
    segmented_img = np.zeros(brain_im.shape, dtype=int) # shape is (256,256)
    
    # Lists
    centering_coord = []
    neighboring_coord = []
    centering_coord.append(seed)
    
    count = 0
    while centering_coord and count < 256**2:
        (x,y) = centering_coord.pop()
        # poping and labelling
        segmented_img[(x,y)] = 1
        
        # resulting neighboring pixel tested with threshold
        neighboring_coord = neighboring_pixels(img, (x,y), threshold).copy()
        
        res = [] 
        for i in neighboring_coord: 
            if i not in res: 
                res.append(i) 
        centering_coord = res.copy()
        count +=1
    return (segmented_img)
```

```python
seed = (125,150)
#seed = (0,0)
threshold = 60
segmented_img = region_growing(brain_im, seed, threshold)

plt.imshow(segmented_img,cmap='gray')
plt.title("Brain area")
plt.show()
```
